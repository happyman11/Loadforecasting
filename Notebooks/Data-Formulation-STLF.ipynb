# Creating a dataset containing hour based analysis, weather detail and Holiday.

Day analysis + Weather + Holiday


```python
import numpy as np
import pandas as pd
```

```python
df_hour = pd.read_excel('/content/1-hour-load-data.xlsx')
df_hour
```

```python
df = pd.read_excel('/content/8-day-load-weather-holiday-data.xlsx')
df
```

```python
df.set_index('date', inplace=True)
df.drop(columns='MW', inplace=True)
```

```python
df
```

```python
df_hour['max-temp'] = 0
df_hour['min-temp'] = 0
df_hour['RH-0830'] = 0
df_hour['RH-1730'] = 0
df_hour
```

```python
# IMPORTANT
# df_hour.loc[2, 'MW'] = 2
```

```python
# Expensive Operation

for index, row in df_hour.iterrows():
  temp_date = str(df_hour.loc[index].DATE)[0:10]
  
  # Extracting field info from "df_hour" dataset using date field

  MAX_TEMP = df.loc[temp_date, 'max-temp']
  MIN_TEMP = df.loc[temp_date, 'min-temp']
  RH1 = df.loc[temp_date, 'RH-0830']
  RH2 = df.loc[temp_date, 'RH-1730']
  DAY = df.loc[temp_date, 'day']

  # Assigning the data in the "df" dataset

  df_hour.loc[index, 'max-temp'] = MAX_TEMP
  df_hour.loc[index, 'min-temp'] = MIN_TEMP
  df_hour.loc[index, 'RH-0830'] = RH1
  df_hour.loc[index, 'RH-1730'] = RH2
  df_hour.loc[index, 'day'] = DAY
```

```python
df_hour
```

```python
df_hour.to_csv('7-hour-load-weather-holiday-data.csv')
```

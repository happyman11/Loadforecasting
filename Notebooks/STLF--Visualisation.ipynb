# Year -  Month - MW


```python
import pandas as pd 
```

```python
df = pd.read_excel('/content/2-day-load-data.xlsx')
df
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import DatetimeTickFormatter
```

```python
x = df['date']
y = df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.line(x, y, legend_label="Visualisation", line_width=2)
p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
```

```python
output_notebook()
show(p)
```

## Convering data to for month-wise analysis with average MW for that month

```python
df.groupby(pd.PeriodIndex(df['date'], freq="M"))['MW'].mean().reset_index()
```

```python
x = df['date']
y = df['MW']

p1 = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p1.line(x, y, legend_label="Visualisation", line_width=2)
p1.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
```

```python
show(p1)
```

```python
x = df['date']
y = df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
p.vbar(x=x,
       top=y,
       legend_label="MW",
       width=0.1,
       bottom=0,
       fill_alpha=0.5,
       color="red",
       alpha=0.5)
```

```python
show(p1)
```

# Holiday, Weekday and Weekend Trends



Convention to be followed:

* 0 - Weekdays (M-F) 
* 1 - Weekends (S, S) 
* 2 - Public Holidays 

```python
import pandas as pd 
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import DatetimeTickFormatter
```

```python
df = pd.read_excel('/content/6-day-load-holiday-data.xlsx')
df
```

### Creating dataset based on type of day

```python
weekday_df = df.loc[df['day'] == 0]
weekend_df = df.loc[df['day'] == 1]
holiday_df = df.loc[df['day'] == 2]
```

## Weekday


```python
x = weekday_df['date']
y = weekday_df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.line(x, y, legend_label="Visualisation", line_width=2)
p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
show(p)
```

```python
x = weekday_df['date']
y = weekday_df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.circle(x, y, size=15, line_color="navy", fill_color="orange", fill_alpha=0.5)
p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
show(p)
```

## Weekend

```python
x = weekend_df['date']
y = weekend_df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.line(x, y, legend_label="Visualisation", line_width=2)
p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
show(p)
```

## Public Holiday

```python
x = holiday_df['date']
y = holiday_df['MW']

p = figure(title="Date vs MW", 
           sizing_mode="stretch_width",
           max_width=900,
           x_axis_type='datetime',
           x_axis_label='Date',
           y_axis_label='MW')

p.line(x, y, legend_label="Visualisation", line_width=2)
p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
show(p)
```
